# **1. What is hypothesis testing in statistics.**

## Hypothesis testing is a formal statistical procedure used to determine whether there is enough evidence in a sample of data to support a particular claim about a population. It's a crucial tool in various fields like science, social sciences, and business to make informed decisions based on data.

# **2. What is the null hypothesis, and how does it differ from the alternative hypothesis.**

## **Null Hypothesis (H0)**
## **Definition**: The null hypothesis is a statement of no effect, no difference, or no relationship between variables.
## **Assumption**: It's the default assumption that researchers start with.
## **Example:**
## If testing a new drug, H0 might be: "The new drug has no effect on the disease."



# **3. What is the significance level in hypothesis testing, and why is it important.**

## The significance level (α) is a crucial concept in hypothesis testing. It's a predetermined threshold that helps us decide whether to reject or fail to reject the null hypothesis.
## **Why is it Important?**
## **Controlling Type I Error:** The significance level directly controls the probability of making a Type I error. A Type I error occurs when we reject the null hypothesis even though it's actually true. By setting a lower α, we reduce the risk of a Type I error.
## **Balancing Risks:** There's a trade-off between Type I and Type II errors. A lower α reduces Type I error but increases the risk of a Type II error (failing to reject a false null hypothesis).
## **Interpreting Results:** The significance level helps us interpret the results of a hypothesis test. A statistically significant result (p-value ≤ α) suggests strong evidence against the null hypothesis.



# **4. What does a P-value represent in hypothesis testing.**

## The p-value is the probability of observing the data (or more extreme data) you obtained from your experiment, assuming the null hypothesis is true
## The p-value is a crucial tool in hypothesis testing. It helps you determine whether your results are statistically significant and provides evidence for or against your research hypothesis. However, it's essential to interpret p-values carefully and consider them within the broader context of your study

# **5. How do you interpret the P-value in hypothesis testing.**

## **Interpreting the P-value**
## **Low P-value (e.g., p < 0.05):** This suggests that it's unlikely you'd have obtained your results if the null hypothesis were true. This provides evidence against the null hypothesis and in favor of the alternative hypothesis.
## **High P-value (e.g., p > 0.05):** This indicates that it's more likely that your results could have occurred by chance if the null hypothesis were true. There's not enough evidence to reject the null hypothesis.

# **6. What are Type 1 and Type 2 errors in hypothesis testing.**

## **Type I and Type II Errors in Hypothesis Testing**

##In hypothesis testing, there are two main types of errors that can occur:

## **Type I Error (False Positive)**

## **Definition:** Rejecting the null hypothesis when it is actually true.
## **Analogy:** Convicting an innocent person in a court of law.
## **Consequence:** Incorrectly concluding that there is a significant effect or difference when there isn't one.
## **Type II Error (False Negative)**

## **Definition:** Failing to reject the null hypothesis when it is actually false.
## **Analogy:** Acquitting a guilty person in a court of law.
## **Consequence:** Failing to detect a true effect or difference.

# **7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing.**

## **One-tailed vs. Two-tailed Tests**

## In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the directionality of your research question.

## **One-tailed Test**

## **Direction:** Predicts a specific direction of the effect (e.g., "Group A will score higher than Group B").1
## **Alternative Hypothesis (H1):**
## H1: Parameter > Value (right-tailed)
## H1: Parameter < Value (left-tailed)
## **Example:**
## A new drug is expected to improve a specific health outcome.
## **Two-tailed Test**

## **Direction:** Predicts a difference but doesn't specify the direction (e.g., "Group A will score differently than Group B").
## Alternative Hypothesis (H1): H1: Parameter ≠ Value
## **Example:**
## Investigating if a new teaching method has any effect on student performance (could be better or worse).

# **8. What is the Z-test, and when is it used in hypothesis testing.**

## A z-test is a statistical hypothesis test used to determine whether the mean of a sample is significantly different from a known population mean.
## **Key Features:**
## Assumes the population standard deviation is known.
## Often used when the sample size is large (generally considered to be 30 or more).
## This is because the Central Limit Theorem states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the underlying population distribution.

## **When is a Z-test Used?**
## Comparing Sample Mean to Population Mean:
## **Example:** Testing whether the average IQ of a group of students is significantly different from the national average IQ.
## **Large Sample Sizes:** When the sample size is sufficiently large, the z-test can be used even if the population standard deviation is unknown (and estimated using the sample standard deviation).

# **9. How do you calculate the Z-score, and what does it represent in hypothesis testing.**

## **Calculating the Z-score**

## The z-score represents the number of standard deviations a data point is from the mean of a distribution. It's calculated using the following formula:

## z = (x - μ) / σ

## **Where:**

##z: The z-score
##x: The data point
##μ: The population mean
##σ: The population standard deviation
## **What does the Z-score represent in Hypothesis Testing?**

## In hypothesis testing, the z-score plays a crucial role in determining whether to reject or fail to reject the null hypothesis. Here's how:

## **Standardization:** The z-score standardizes the data, allowing us to compare values from different distributions.
## **Comparison to Critical Values:** The calculated z-score is compared to critical values from the standard normal distribution table. These critical values are determined based on the significance level (α) and the type of test (one-tailed or two-tailed).
## **Decision Making:** If the absolute value of the calculated z-score is greater than the critical value, it falls in the rejection region. This means we have enough evidence to reject the null hypothesis.
## If the absolute value of the calculated z-score is less than or equal to the critical value, it falls within the acceptance region. We fail to reject the null hypothesis.



# **10. What is the T-distribution, and when should it be used instead of the normal distribution.**

## **Definition:** The t-distribution is a probability distribution that is similar to the normal distribution but has heavier tails. This means it's more likely to produce extreme values than a normal distribution.
## **Key Features:**
## **Bell-shaped:** Like the normal distribution, it's symmetrical and bell-shaped.
## **Heavier Tails:** It has more probability in the tails compared to the normal distribution.
## **Degrees of Freedom:** The shape of the t-distribution is determined by its degrees of freedom, which are related to the sample size.
## **When to Use the T-distribution Instead of the Normal Distribution**

## You should use the t-distribution instead of the normal distribution when:

## **Unknown Population Standard Deviation:** The most crucial condition. When you don't know the population standard deviation and have to estimate it using the sample standard deviation, the t-distribution is more appropriate.
## **Small Sample Size:** The t-distribution is particularly useful when the sample size is small (typically less than 30). With small samples, the sample standard deviation may not accurately reflect the population standard deviation, and the t-distribution accounts for this uncertainty.

# **11. What is the difference between a Z-test and a T-test.**

## **Z-test:** A more specific test used when you have a known population standard deviation and often with larger sample sizes.
## **T-test:** A more general and widely used test, particularly suitable when the population standard deviation is unknown, especially with smaller sample sizes.

# **12. What is the T-test, and how is it used in hypothesis testing.**

## **Definition:** A t-test is a statistical hypothesis test used to compare the means of two groups. It's a common tool used in various fields to determine if there's a significant difference between the average values of two sets of data.
## **How T-tests are Used in Hypothesis Testing**

## **State the Hypotheses:**

## **Null Hypothesis (H0):** There is no significant difference between the means of the two groups.
## **Alternative Hypothesis (H1):** There is a significant difference between the means of the two groups. (This can be one-tailed or two-tailed, as discussed earlier).
## **Calculate the T-statistic:**

## The t-statistic is calculated based on the difference between the sample means, the standard deviation of the samples, and the sample sizes.
## **Determine Degrees of Freedom:**

##The degrees of freedom depend on the type of t-test and sample sizes.
## **Find the Critical Value:**

##Using the t-distribution table and degrees of freedom, find the critical t-value for the chosen significance level (e.g., 0.05).
## **Compare Calculated T-statistic to Critical Value:**

## If the calculated t-statistic is greater than the critical value (in absolute terms), reject the null hypothesis.
## If the calculated t-statistic is less than or equal to the critical value, fail to reject the null hypothesis.

# **14. What is a confidence interval, and how is it used to interpret statistical results.**

## **Definition:** A confidence interval (CI) is a range of values that is likely to contain the true value of an unknown population parameter.
## **Key Concepts:**
## **Population Parameter:** The true value of a characteristic within an entire population (e.g., the true average height of all adults in a country).
## **Sample Statistic:** An estimate of the population parameter calculated from a sample of data (e.g., the average height of a sample of 100 adults).
## **Confidence Level:** The probability that the interval will contain the true population parameter. Common levels are 95% and 99%.
## **How Confidence Intervals are Used to Interpret Statistical Results**

## **Quantifying Uncertainty:** Confidence intervals provide a measure of the uncertainty associated with a sample statistic. A wider interval indicates greater uncertainty, while a narrower interval suggests more precision.
## **Making Inferences:**
## If a confidence interval for the difference between two groups does not include zero, it suggests that there is a statistically significant difference between the groups.
## If a confidence interval for a population mean includes a specific value, it suggests that the true population mean may be that value.
## **Communicating Results:** Confidence intervals provide a more informative way to communicate research findings than simply reporting p-values. They give readers a better understanding of the range of plausible values for the population parameter.

# **16. How is Bayes' Theorem used in statistics, and what is its significance.**

## Bayes' Theorem is a fundamental principle in probability and statistics that describes how to update the probability of an event based on new evidence. In simpler terms, it helps us revise our beliefs about the likelihood of something happening as we gather more information.

## The Formula

## **P(A|B) = [P(B|A) * P(A)] / P(B)**
## **Where:**

## **P(A|B):** The probability of event A occurring given that event B has already occurred. This is called the posterior probability.
## **P(B|A):** The probability of event B occurring given that event A has already occurred. This is often called the likelihood.
## **P(A):** The prior probability of event A occurring. This is our initial belief about the probability of A before considering any new evidence.
## **P(B):** The probability of event B occurring.
## **How Bayes' Theorem is Used in Statistics**

## **Updating Beliefs:** Bayes' Theorem allows us to update our prior beliefs about the probability of an event based on new evidence. This is crucial in many areas, such as:
## **Machine learning:** Used in algorithms like Bayesian networks and Naive Bayes classifiers for making predictions and classifications.
## **Medical diagnosis:** Helps doctors revise the probability of a disease based on new test results.
## **Artificial intelligence:** Used in various AI applications, such as spam filtering and fraud detection.
## **Scientific research:** Used to update theories and hypotheses based on new experimental data.
## **Significance of Bayes' Theorem**

## **Foundation for Bayesian Inference:** Bayes' Theorem is the foundation of Bayesian statistics, a powerful approach to statistical inference that focuses on updating beliefs in the light of new evidence.
## **Flexibility:** It can be applied to a wide range of problems, from simple decision-making to complex machine learning models.
## **Handles Uncertainty:** Bayes' Theorem provides a principled way to deal with uncertainty and incorporate prior knowledge into our analysis.

## **Chi-Square Test of Independence:**

## Can be used to test hypotheses about the variance of a normally distributed population.




# **18. What is the Chi-square goodness of fit test, and how is it applied.**

## **Chi-Square Goodness-of-Fit Test**

## **Purpose:** This test determines how well observed data fits a particular expected distribution. In simpler terms, it checks if your sample data comes from a population with a specific distribution (like a uniform distribution, a normal distribution, or a distribution you've hypothesized).
## **How it Works:**

## **Define Hypotheses:**

## **Null Hypothesis (H0):** The observed data fits the expected distribution.
## **Alternative Hypothesis (H1):** The observed data does not fit the expected distribution.
## **Calculate Expected Frequencies:**

## Determine the expected frequencies for each category based on the hypothesized distribution.
## **Calculate the Chi-Square Statistic:**

## This involves comparing the observed frequencies to the expected frequencies using the following formula:

## **χ² = Σ [(Observed Frequency - Expected Frequency)² / Expected Frequency]**

## **Where:**

## **χ²:** The chi-square test statistic
## **Σ:** Sum of all categories
## **Determine Degrees of Freedom:**

## Degrees of freedom (df) = Number of categories - 1
## **Find the Critical Value:**

## Use a chi-square distribution table to find the critical value based on the chosen significance level (e.g., 0.05) and the degrees of freedom.
## **Compare Calculated and Critical Values:**

## If the calculated chi-square value is greater than the critical value, reject the null hypothesis. This means the observed data significantly differs from the expected distribution.
## If the calculated chi-square value is less than or equal to the critical value, fail to reject the null hypothesis.

# **3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.**

```python
from scipy.stats import norm
import numpy as np

def one_sample_z_test(sample_data, population_mean, population_std, alfa = 0.05):
  sample_mean = np.mean(sample_data)
  sample_size = len(sample_data)
  z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
  p_value = 2 * (1 - norm.cdf(abs(z_score)))
  return z_score, p_value

sample_data = [102, 98, 105, 97, 103, 101, 99, 104, 100, 98]
population_mean = 100
population_std = 5
alfa = 0.05
z_score, p_value = one_sample_z_test(sample_data, population_mean, population_std)
print(f"z-score: {z_score}")
print(f"p-value: {p_value}")
if p_value < alfa:
  print("Reject the null hypothesis")
  print("There is sufficient evidence to suggest that the true population mean is different from the hypothesized mean.")
else:
  print("Fail to reject the null hypothesis")
  print("There is sufficient evidence to suggest that the true population mean is different from the hypothesized mean.")
```

# **4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_data, population_mean, population_std, alfa = 0.05):

  sample_mean = np.mean(sample_data)
  sample_size = len(sample_data)

  z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
  p_value = 2 * norm.sf(abs(z_score))
  return z_score, p_value

sample_data = [102, 98, 105, 97, 103, 101, 99, 104, 100, 98]
population_mean = 100
population_std = 5
alfa = 0.05

z_score, p_value = two_tailed_z_test(sample_data, population_mean, population_std)
z_critical = norm.ppf(1 - alfa/2)
z_critical_left = -z_critical

x = np.linspace(-4, 4, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, norm.pdf(x), label='Standard Normal Distribution')
plt.fill_between(x, norm.pdf(x), where=x >= z_critical_left, color='red', alpha=0.2, label= 'Decision Region')
plt.fill_between(x, norm.pdf(x), where=x <= z_critical, color='red', alpha=0.2)
plt.axvline(x=z_critical_left, color='red', linestyle='--', label='Critical Value')
plt.axvline(x=z_critical,color='red', linestyle='--')
plt.xlabel('Z-Score')
plt.ylabel('Probabilty Density')
plt.title('Two_tailed_Z_test')
plt.legend()
print(f"z-score: {z_score}")
print(f"p-value: {p_value}")

if p_value < alfa:
  print("Reject the null hypothesis")
  print("There is sufficient evidence to suggest that the true population mean is different from the hypothesized mean.")
else:
  print("Fail to reject the null hypothesis")
  print("There is not sufficient evidence to suggest that the true population mean is different from the hypothesized mean")

plt.show()
```

# **5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type_i_ii_error(null_mean, true_mean, std_dev,sample_size, alpha = 0.05):

  # Calculate standard error
  standard_error = std_dev / np.sqrt(sample_size)

  # Calculate critical values for Type I error
  z_critical = norm.ppf(1 - alpha/2)
  critical_value_upper = null_mean + z_critical * standard_error
  critical_value_lower = null_mean - z_critical * standard_error

  # Define x-axis range
  x = np.linspace(null_mean - 4 * std_dev, true_mean + 4 * std_dev, 1000)

  # Calculate probability density functions
  null_dist = norm.pdf(x, loc = null_mean, scale = standard_error)
  true_dist = norm.pdf(x, loc = true_mean, scale = standard_error)

  # Plot distributions
  plt.figure(figsize=(10, 6))
  plt.plot(x, null_dist, label=f' Null Hypothesis Distribution (μ={null_mean})' , color='blue')
  plt.plot(x, true_dist, label=f'True Hypothesis Distribution (μ={true_mean})' , color='red')

  # Shade Type I error region
  plt.fill_between(x,null_dist, where=(x > critical_value_upper) | (x < critical_value_lower), color='red', alpha=0.02, label='Type I Error Region')

  # Shade Type II error region
  # (Approximate: Assuming critical values are the same for both distributions)
  plt.fill_between(x, true_dist, where=(x > critical_value_lower) | (x < critical_value_upper), color = 'green', alpha = 0.02, label = 'Type II Error Region')

  # plot critical values
  plt.axvline(critical_value_upper, color='red', linestyle='--', label='Critical_value')
  plt.axvline(critical_value_lower, color='red', linestyle='--')
  plt.xlabel('Sample_mean')
  plt.ylabel('Probability Density')
  plt.title('Type I and II Error Visualization')
  plt.legend()
  plt.show()

null_mean = 100
true_mean = 105  # Alternative hypothesis: True mean is different from null mean
std_dev = 15
sample_size = 30
alpha = 0.05

visualize_type_i_ii_error(null_mean, true_mean, std_dev,sample_size, alpha)
```

# **6. Write a Python program to perform an independent T-test and interpret the results.**

```python
import numpy as np
from scipy.stats import ttest_ind, t

def independent_t_test(data1, data2, alpha):

  t_stat, p_value = ttest_ind(data1,data2)
  return t_stat, p_value

data1 = [10, 12, 15, 18, 20]
data2 = [8, 11, 13, 14, 16]
alpha = 0.05

t_stat, p_value = independent_t_test(data1,data2, alpha)

print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < alpha:
  print("Reject the null hypothesis.")
  print("There is sufficient evidence to suggest that the means of the two groups are significantly different.")
else:
  print("Fail to reject the null hypothesis.")
  print("There is not enough evidence to suggest that the means of the two groups are significantly different.")
```

# **6.Perform a paired sample T-test using Python and visualize the comparison results.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel, t

def paired_t_test(data1, data2, alpha = 0.05):
  t_stat, p_value = ttest_rel(data1, data2)
  return t_stat, p_value

data1 = [10, 12, 15, 18, 20]
data2 = [8, 11, 13, 14, 16]

t_stat, p_value = paired_t_test(data1, data2)
df = len(data1)-1

critical_value = t.ppf(1-alpha/2, df)

x = np.linspace(-4, 4, 100)

plt.figure(figsize=(8, 4))
plt.plot(x, t.pdf(x, df), label=f't-distribution (df={df})')

plt.fill_between(x, t.pdf(x, df), where=(x < -critical_value), color='red', alpha=0.2, label='Decision Region')
plt.fill_between(x, t.pdf(x, df), where=(x > critical_value), color='red', alpha=0.2)

plt.axvline(x=-critical_value, color='red', linestyle='--', label='Critical Value')
plt.axvline(x=critical_value, color='red', linestyle='--')

plt.xlabel('t-statistics')
plt.ylabel('Probability Density')
plt.title('Paired_t_test')
plt.legend()

print("t-statistic:", t_stat)
print("p-value:", p_value)


if p_value < alpha:
  print("Reject the null hypothesis.")
  print("There is sufficient evidence to suggest that the means of the two groups are significantly different.")
else:
  print("Fail to reject the null hypothesis.")
  print("There is not enough evidence to suggest that the means of the two groups are significantly different.")

plt.show()
```

# **9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):

  n = len(data)
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

  # Calculate the critical value (t-score)
  t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

  # Calculate the margin of error
  margin_of_error = t_critical * (sample_std / np.sqrt(n))

  # Calculate the confidence interval
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error

  return lower_bound, upper_bound

# Example usage
data = [10, 12, 15, 18, 20, 17, 14, 19, 16, 11]
confidence_level = 0.95  # 95% confidence interval

lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

print(f"Confidence Interval ({confidence_level*100}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
```

# **24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results**

```python
import numpy as np
from scipy.stats import chi2

def variance_hypothesis_test(sample, hypothesized_variance, alpha=0.05):
    """
    Perform a hypothesis test for population variance using the Chi-square distribution.

    Parameters:
        sample (array-like): The sample data.
        hypothesized_variance (float): The hypothesized population variance.
        alpha (float): Significance level (default: 0.05).

    Returns:
        None: Prints the test statistic, critical values, and conclusion.
    """
    # Sample statistics
    n = len(sample)
    sample_variance = np.var(sample, ddof=1)

    # Chi-square test statistic
    chi_square_stat = (n - 1) * sample_variance / hypothesized_variance

    # Degrees of freedom
    df = n - 1

    # Critical values
    chi_square_critical_low = chi2.ppf(alpha / 2, df)
    chi_square_critical_high = chi2.ppf(1 - alpha / 2, df)

    # Print results
    print("Hypothesis Test for Population Variance:")
    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Test Statistic (Chi-square): {chi_square_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"Critical Values: ({chi_square_critical_low:.4f}, {chi_square_critical_high:.4f})")

    # Conclusion
    if chi_square_stat < chi_square_critical_low or chi_square_stat > chi_square_critical_high:
        print("Conclusion: Reject the null hypothesis. The variance is significantly different from the hypothesized variance.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. The variance is not significantly different from the hypothesized variance.")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, SD=10, Size=30
hypothesized_variance = 100  # Hypothesized population variance

variance_hypothesis_test(sample_data, hypothesized_variance)
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(success1, n1, success2, n2, alpha=0.05):
    """
    Perform a Z-test to compare proportions between two groups.

    Parameters:
        success1 (int): Number of successes in group 1.
        n1 (int): Total number of observations in group 1.
        success2 (int): Number of successes in group 2.
        n2 (int): Total number of observations in group 2.
        alpha (float): Significance level (default: 0.05).

    Returns:
        None: Prints the Z-statistic, p-value, and conclusion.
    """
    # Combine success and total observations
    successes = np.array([success1, success2])
    totals = np.array([n1, n2])

    # Perform Z-test
    stat, p_value = proportions_ztest(successes, totals)

    # Print results
    print("Z-Test for Comparing Proportions:")
    print(f"Z-statistic: {stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis. The proportions are significantly different.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. The proportions are not significantly different.")

# Example usage
success1 = 40  # Number of successes in group 1
n1 = 100       # Total observations in group 1
success2 = 30  # Number of successes in group 2
n2 = 90        # Total observations in group 2

z_test_proportions(success1, n1, success2, n2)
```

#  **26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results**

```python
import numpy as np
from scipy.stats import f

def f_test_variances(data1, data2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two datasets.

    Parameters:
        data1 (array-like): Dataset 1.
        data2 (array-like): Dataset 2.
        alpha (float): Significance level (default: 0.05).

    Returns:
        None: Prints the test statistic, critical values, and conclusion.
    """
    # Sample variances and sizes
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    n1, n2 = len(data1), len(data2)

    # Calculate F-statistic
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    dfn = n1 - 1  # Degrees of freedom for numerator
    dfd = n2 - 1  # Degrees of freedom for denominator

    # Critical values
    f_critical_low = f.ppf(alpha / 2, dfn, dfd)
    f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

    # Print results
    print("F-Test for Comparing Variances:")
    print(f"Variance 1: {var1:.4f}, Variance 2: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"Degrees of Freedom (Numerator, Denominator): ({dfn}, {dfd})")
    print(f"Critical Values: ({f_critical_low:.4f}, {f_critical_high:.4f})")

    # Conclusion
    if f_stat < f_critical_low or f_stat > f_critical_high:
        print("Conclusion: Reject the null hypothesis. The variances are significantly different.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. The variances are not significantly different.")

    # Visualization
    import matplotlib.pyplot as plt
    import seaborn as sns

    plt.figure(figsize=(10, 6))
    sns.kdeplot(data1, label='Dataset 1', fill=True, color='blue', alpha=0.5)
    sns.kdeplot(data2, label='Dataset 2', fill=True, color='orange', alpha=0.5)
    plt.axvline(np.mean(data1), color='blue', linestyle='--', label=f'Mean Dataset 1: {np.mean(data1):.2f}')
    plt.axvline(np.mean(data2), color='orange', linestyle='--', label=f'Mean Dataset 2: {np.mean(data2):.2f}')
    plt.title("Distributions of Two Datasets")
    plt.xlabel("Values")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Example usage
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, SD=10, Size=30
data2 = np.random.normal(loc=55, scale=15, size=30)  # Mean=55, SD=15, Size=30

f_test_variances(data1, data2)
```

# **27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.**

```python
import numpy as np
from scipy.stats import chisquare

def simulate_data(num_categories, num_samples, probabilities=None):
  """
  Simulates categorical data with specified probabilities.

  Args:
    num_categories: Number of categories.
    num_samples: Total number of samples.
    probabilities: A list of probabilities for each category.
                   If None, assumes a uniform distribution (equal probabilities).

  Returns:
    A list of observed frequencies for each category.
  """

  if probabilities is None:
    probabilities = [1/num_categories] * num_categories

  # Generate random categorical data
  data = np.random.choice(range(num_categories), size=num_samples, p=probabilities)

  # Count the occurrences of each category
  observed_frequencies = np.bincount(data)

  return observed_frequencies

def perform_chi_square_goodness_of_fit_test(observed_frequencies, expected_frequencies):
  """
  Performs a Chi-Square goodness-of-fit test.

  Args:
    observed_frequencies: A NumPy array of observed frequencies.
    expected_frequencies: A NumPy array of expected frequencies.

  Returns:
    chi_square_stat: The calculated Chi-Square statistic.
    p_value: The p-value of the test.
  """

  chi_square_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)
  return chi_square_stat, p_value

# Simulation parameters
num_categories = 4
num_samples = 100
probabilities = [0.25, 0.3, 0.2, 0.25]  # Example: Unequal probabilities

# Simulate data
observed_frequencies = simulate_data(num_categories, num_samples, probabilities)

# Calculate expected frequencies (assuming equal probabilities in this case)
expected_frequencies = np.full(num_categories, num_samples / num_categories)

# Perform Chi-square test
chi_square_stat, p_value = perform_chi_square_goodness_of_fit_test(observed_frequencies, expected_frequencies)

# Print results
print("Observed Frequencies:", observed_frequencies)
print("Expected Frequencies:", expected_frequencies)
print("Chi-Square Statistic:", chi_square_stat)
print("p-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis.")
  print("There is evidence that the observed data does not fit the expected distribution.")
else:
  print("Fail to reject the null hypothesis.")
  print("There is not enough evidence to conclude that the observed data differs from the expected distribution.")
```
